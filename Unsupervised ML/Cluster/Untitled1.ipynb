```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

df = pd.read_csv("cats_vs_dogs_features.csv")
df.head()
```

```python
X = df.drop(columns=['label'])
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X)

df['cluster'] = kmeans.labels_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['weight (kg)'], df['height (cm)'], c=df['cluster'], cmap='Set1', alpha=0.7)
plt.xlabel('Poids (kg)')
plt.ylabel('Taille (cm)')
plt.title('Clusters détectés par K-Means')
plt.grid(True)
plt.show()
```

```python

pd.crosstab(df['label'], df['cluster'], rownames=['Réel'], colnames=['Cluster trouvé'])
```
